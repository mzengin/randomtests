```python
import numpy as np
```

```python
class LinearRegression:
  def __init__(self):
    pass


  def fit(self, X, y, lr= 0.01, iteration=500):
    #  X is m X n
    #  y is m X 1
    m = X.shape[0]
    n = X.shape[1]

    #  theta is 1 by n+1
    self.theta = np.random.random_sample((1, n + 1))
    self.lr = lr
    # X is m by n+1
    X = np.hstack((X, np.ones((m, 1))))
    for i in range(iteration):

      # pred is m X 1
      pred = np.dot(X, self.theta.T)

      mse = 1/m * np.mean((pred - y) ** 2)
      print(f"iteration {i}, mse {mse}")

      # gradient is 1 X n+1
      gradient = 2/m * np.dot((pred-y).T, X)
      self.theta = self.theta - lr * gradient

    print(self.theta)


```

```python
X = np.random.random_sample((50, 1)) * 10
```

```python
y = X + np.random.random_sample((50, 1)) * 5
```

```python
import matplotlib.pyplot as plt
```
